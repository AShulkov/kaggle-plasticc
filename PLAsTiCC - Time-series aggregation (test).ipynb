```python
col_list = ['flux_min_pb0','flux_min_pb1','flux_min_pb2','flux_min_pb3','flux_min_pb4','flux_min_pb5',
            'flux_max_pb0','flux_max_pb1','flux_max_pb2','flux_max_pb3','flux_max_pb4','flux_max_pb5',
            'flux_mean_pb0','flux_mean_pb1','flux_mean_pb2','flux_mean_pb3','flux_mean_pb4','flux_mean_pb5',
            'flux_std_pb0','flux_std_pb1','flux_std_pb2','flux_std_pb3','flux_std_pb4','flux_std_pb5',
            'flux_skew_pb0','flux_skew_pb1','flux_skew_pb2','flux_skew_pb3','flux_skew_pb4','flux_skew_pb5',
            'flux_range_pb0','flux_range_pb1','flux_range_pb2','flux_range_pb3','flux_range_pb4','flux_range_pb5',
            'flux_relrange_pb0','flux_relrange_pb1','flux_relrange_pb2','flux_relrange_pb3','flux_relrange_pb4','flux_relrange_pb5'
            ]
```

```python
%%time
df_tail = pd.DataFrame()
chunksize = 10 ** 6
for i, df_chunk in enumerate(pd.read_csv('test_set.csv', sep = ',', header = 0, chunksize=chunksize, 
                            dtype = {'passband': np.object})
                         ):
    print(i)
    df_chunk.passband = 'passband_'+ df_chunk.passband
    df_chunk.passband = df_chunk.passband.astype('category')
    print (df_chunk.shape)
    last_obj_id = df_chunk.object_id.iloc[-1] 
    df_main = pd.concat((df_tail, df_chunk[df_chunk.object_id != last_obj_id]))
    df_tail = df_chunk[df_chunk.object_id == last_obj_id]
    print (df_main.shape, df_tail.shape)
    #print(df_main.head())
    
    df_temp = df_main.groupby(['object_id', 'passband']).flux.agg({'min', 'max', 'mean', 'std', 'skew'})
    df_temp = df_temp[['min', 'max', 'mean', 'std', 'skew']]
    df_temp.reset_index(level = 1, drop = False, inplace = True)
    df_temp['flux_range'] = df_temp['max'] - df_temp['min']
    df_temp['flux_rel_range'] = df_temp['flux_range']/df_temp['mean']
    print(df_temp.shape)
    
    df_grouped = df_temp.pivot(columns = 'passband')
    #df_grouped.columns = col_list
    print(df_grouped.shape)
    df_grouped.to_csv('df_test_FE.csv', index = True, header = False, mode='a')
    del df_main, df_temp, df_grouped
    gc.collect()
    print(datetime.now())
```
